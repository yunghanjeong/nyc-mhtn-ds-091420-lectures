```python
# Let's get started importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
# Loading the data
df = pd.read_csv("ames_train.csv")
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['LotArea', 'SalePrice', 'GarageArea', 'YrSold', 'YearBuilt']], figsize=(15,15));
```

```python
# Create a plot that shows the SalesPrice Distribution
fig, ax = plt.subplots(figsize=(6,6))
ax.hist(df["SalePrice"], bins = 8)
ax.set_title("Sale Price Distribution")
ax.set_xlabel("Sale Price")
ax.set_ylabel("Sale Count")
ax.axvline(df['SalePrice'].mean(), color='red'); #this is very cool
```

```python
# Create a plot that shows the LotArea Distribution
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(df["LotArea"], bins = "auto")
ax.set_title("Lot Area Distribution")
ax.set_xlabel("Lot Area")
ax.set_ylabel("Distribution")
ax.axvline(df['LotArea'].mean(), color='purple'); #this is very cool
```

```python
# Create a plot that shows the Distribution of the overall house condition
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(df["OverallCond"], bins = "auto")
ax.set_title("Property Overall Condition Distribution")
ax.set_xlabel("Condition (10- Best, 1-Worst)")
ax.set_ylabel("Property Count")
```

```python
# Create a Box Plot for SalePrice
fig, ax = plt.subplots(figsize=(4,6))
ax.boxplot(df["SalePrice"])
ax.set_title("Sale Price Box Distribution")
```
