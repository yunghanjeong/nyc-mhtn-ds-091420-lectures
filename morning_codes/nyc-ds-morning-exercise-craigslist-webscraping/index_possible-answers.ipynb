```python
def info_grabber(url):
    
    # convert page to soup object
    page = rq.get(url)
    soup = bs(page.content, 'html.parser')
    
    # instantiate empty dictionary
    info = {}
    
    # scrape price
    info['price'] = float(soup.find('span', attrs={'class':'price'}).contents[0]\
                            .replace('$', '').replace(',', ''))
    
    # scrape location
    try:
        info['location'] = soup.find_all('small')[0].contents[0].strip()\
                                .replace('(', '').replace(')', '')\
                                .replace(' ', '_').replace('-', '_')\
                                .lower()
    # track any posts without a location
    except IndexError:
        print(f'Location Error: {url}')
        # put NaN in its place
        info['location'] = np.nan
    
    # bedrooms
    info['num_beds'] = int(soup.find('span', attrs={'class':'shared-line-bubble'})\
                               .contents[0].contents[0][:-2])
    
    # bathrooms
    try:
        info['num_baths'] = float(soup.find('span', attrs={'class':'shared-line-bubble'})\
                                    .contents[2].contents[0][:-2])
    # track any posts "without" a bathroom
    # generally has a value of "shared"
    except ValueError:
        print(f'Bathroom Error: {url}')
        # put NaN in its place
        info['num_baths'] = np.nan
    
    # include craigslist url, for reference
    info['url'] = url
    
    return info
```

```python
%%time

# grab info for each link/apartment
apts_info = [info_grabber(link) for link in links]

# look at first example
apts_info[0]
```
